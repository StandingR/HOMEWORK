## 2.분산(편차의 제곱) 

분산은 편차(개별값-평균값)를 제곱한 후 다 더해서 전체 자료의 수로 나누어 준 값이다.<br>
즉, <span style="color:red">편자의 제곱합을 평균</span>한 값이다.

이러한 방법으로 하면 편차의 합이 0이 되는 것을 막을 수 있다.<br>분산을 수식으로 나타내면 다음과 같다.
![스크린샷 2023-09-20 18.45.52.png](<attachment:스크린샷 2023-09-20 18.45.52.png>)

분산은 편차의 제곱을 평균한 값이기 떄문에 자료의 단위 역시 제곱이 된다.<br>
예를 들어 학생들의 키의 분산을 계산하면 그 단위는 길이 단위인 cm가 아니라 면적의 단위인 cm**2이 된다.<br>
이렇게 단위가 달라지게 되면 평균적으로 어느 정도 편차가 있는 건지 알 수가 없다.<br>이를 해결하기 위해 나온 것이 표준편차이다.

## 3.표쥰편차(분산 값에 제곱근(루트)을 취한값->하는 이유는 원래의 단위로 바꿔주기 위해)

표준편차는  <span style="color:red">분산 값에 제곱근을 취한 값</span>이다.<br>이렇게 분산에 제곱근을 취하게 되면 단위가 원래 자료의 단위로 환원하게 된다.<br>
쉽게 말해서 -> 분산의 단위를 원래의 자료의 단위에 맞게 전환해서 이해하기 쉽게 만든 값이 표준편차라고 할수있다.
![스크린샷 2023-09-20 20.08.41.png](<attachment:스크린샷 2023-09-20 20.08.41.png>)

```python
# 1- 13
#뷴산과 표준편차를 이용하여 수학식(standardization)을 코드로 만드는 연습을 해봅니다.

score1 = 10
score2 = 20
score3 = 30
n_student = 3

# 제평 - 평제하기 = 분산(variacne[veə.ri.əns/베리언스])
score_mean = (score1 + score2 +score3)/n_student
sqaure_of_mean = score_mean**2
mean_of_square = (score1**2 + score2**2 + score3**2)/n_student

score_variance = mean_of_square - sqaure_of_mean 
score_std = score_variance**0.5


print("mean: ",score_mean)
print("standard deviation: ", score_std) #standard deviation(Std) = 표준편차

score1 = (score1 - score_mean)/score_std
score2 = (score2 - score_mean)/score_std
score3 = (score3 - score_mean)/score_std

score_mean = (score1 + score2 + score3)/n_student
square_of_mean = score_mean**2 # mean
mean_of_square = (score1**2 + score2**2 + score3**2)/n_student
score_variance = mean_of_square - square_of_mean
score_std = score_variance**0.5

print("mean: ", score_mean)
print("standrad deviation: ", score_std)

```

표준화는 어떤 특성의 값들이 정규분포,<br>즉 종모양의 분포를 따른다고 가정하고 값들을 0의 평균,1의 표준편차를 갖도록 변환해주는 것
