Vector Norm - Vector의 크기를 알고 싶을 경우에는 Norm을 사용합니다.<br>

즉, Vector의 크기를 재어주는 함수를 Norm이라고 합니다.(Euclidean norm)

수학식(vector norm)을 코드로 만드는 연습

Vector Norm은 간단하게 말하면 두점 사이의 거리를 구한다라고 생각한다.
이는 피타고라스의 정리로 증명할수있다.

이를 하는 이유는 Unit Vector를 만들어 주기 위함이다.

```python
# EX 1-16 Vector Norm
# 수학식 (vector norm)을 코드로 만드는 연습을 해봅니다.
# 0.5는 루트

x,y,z = 1, 2, 3

norm = (x**2 + y**2 + z**2)**0.5
print(norm)
```

### 유닛벡터 - https://velog.io/@yuns_u/%EB%8B%A8%EC%9C%84%EB%B2%A1%ED%84%B0-%EA%B8%B0%EC%A0%80%EB%B2%A1%ED%84%B0-span-rank-linear-projections(유닛벡터의 의미)

단위 벡터(Unit Vactors) = 단위 길이 (1)를 갖는 모든 벡터



단위 백터는 다양하지만, 특히 단위 백터의 element 중 하나만 1이고 나머지가 0인 것들을 차원의 단위백터라고도 한다.


```python
# EX 1-17 Making UNit Vectors
# 수학식 (unit vextor를 만드는 방법)을 코드로 만드는 연습을해봅니다.

x,y,z =1,2,3

# 벡터 놈은 각 벡터의 거리
norm = (x**2 + y**2 + z**2)**0.5
print(norm)

# Unit Vector는 각 벡터를 norm으로 나눈것. 그 후에 다시 제곱해서 루트를 씌워주면 1로 변환된다.
x,y,z = x/norm,y/norm,z/norm # 유닛백터 완료
norm = (x**2 + y**2 + z**2)**0.5 #유닛백터  
print(norm)
```

```python
# Exercise 1-19 : Euclidean distance
# 수학식 (Euclidean distance)를 코드로 만드는 연습을 해봅니다

x1, y1, z1 = 1, 2, 3
x2, y2, z2 = 3, 4, 5

e_distacne = (x1 - x2)**2 + (y1 - y2)**2 + (z1 - z2)**2
e_distacne **= 0.5

print(e_distacne)
```

두 포인트 사이의 유클리디언 거리는 구하는 방법은 각각의 동일 차원 축 (x, y...) 간 거리를 뺀 값을 제곱하고 모두 더한 뒤 루트를 씌워준다 (하기 공식 참고)

MSE(Mean Squared Error)

MSE란 Mean Squared Error의 약자로, 예측 값과 실제 값의 차이에 대한 제곱을 평균한 값을 말합니다.<br>
MSE의 경우 오차가 작으면 성늠이 좋을 수있지만, 과대적합(통계 모델이 학습 데이터와 정확히 일치하는 경우)이 될 수도 있습니다.<br>
한편, MSE는 오차에 대해서 예측한 값이 실제 값보다 큰지, 작은지를 알 수 없다는 특징을 갖고 있습니다.


```python
# EX 1- 21 : Mean Squared Error
# 수학식 (MSE을 코드로 만드는 연습을 해봅니다.
pred1, pred2, pred3 = 10, 20, 30
y1, y2, y3 = 10, 25, 40
n_data = 3

s_error1 = (pred1 - y1)**2
s_error2 = (pred2 - y2)**2
s_error3 = (pred3 - y3)**2

mean_squared_Error = (s_error1 + s_error2 + s_error3)/n_data

print(mean_squared_Error)
```
